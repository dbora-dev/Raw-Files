```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

import tensorflow_datasets as tf_data
import matplotlib.pyplot as plt
```

```python
# Preprocess the MNIST images :
```

```python
def map_image(images,label): 
    # Normalize the image pixels and flatten the image 
    image = tf.cast(images,dtype=tf.float32)
    image = image/255.0
    image = np.reshape(image,shape=(28*28*1,))
    label = np.reshape(image,shape=(28*28*1,))
    
    return image, label




```

```python
# Split the training the Testing set :
```

```python
BATCH_SIZE = 128
BUFFER_SIZE = 1024
```

```python
train_dataset = tf_data.load("mnist",as_supervised=True,split='train')
train_dataset = train_dataset.map(map_image)
train_dataset = train_dataset.shuffle(BATCH_SIZE).batch(BATCH_SIZE).repeat()

test_datasets = tf_data.load('mnist',as_supervised=True,split='test')
test_datasets = test_datasets.map(map_image)
test_datasets = test_datasets.batch(BATCH_SIZE).repeat()
```
