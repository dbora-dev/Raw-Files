```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.model_selection import train_test_split
```

```python
seed = 1
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
x = pd.read_csv("./avila-tr_feats.csv")
y = pd.read_csv("./avila-tr_target.csv")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=seed)
```

```python
# define a deep Learning model
model = Sequential()
model.add(Dense(units=10,input_dim=x_train.shape[1],activation='relu'))
model.add(Dense(units=6,activation='relu'))
model.add(Dense(units=4,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
# Train the model on training sets and Evaluate the model on Test sets
history = model.fit(x=x_train,y=y_train,epochs=100,batch_size=20,validation_data=(x_test,y_test),verbose=1,shuffle=False)
```

```python
print(f"Max Training Loss = {max(history.history['loss'])}.\n Max Validation Loss or Test Loss = {max(history.history['val_loss'])} \n Max accuracy = {max(history.history['accuracy'])}")
```

```python
%matplotlib inline
```

```python
mpl.rcParams['figure.figsize'] = (16.0,8.0)
plt.ylim(0,1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(['Training Loss','Validation Loss'],loc='upper right')

# Max acuuracy reached on Test or Validation Set : 
print(f"Max acuracy reached on Test or Validation Set : {max(history.history['val_accuracy'])}")
```

```python
from keras.regularizers import l1,l2,l1_l2
```

```python
l2_param =0.001
```

```python
# define a deep Learning model
model_1 = Sequential()
model_1.add(Dense(units=10,input_dim=x_train.shape[1],activation='relu',kernel_regularizer=l2(l2_param)))
model_1.add(Dense(units=6,activation='relu',kernel_regularizer=l2(l2_param)))
model_1.add(Dense(units=4,activation='relu',kernel_regularizer=l2(l2_param)))
model_1.add(Dense(units=1,activation='sigmoid'))
model_1.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
history_1 = model_1.fit(x=x_train,y=y_train,epochs=100,batch_size=20,validation_data=(x_test,y_test),verbose=1,shuffle=False)
```

```python
mpl.rcParams['figure.figsize'] = (16.0,8.0)
plt.ylim(0,1)
plt.plot(history_1.history['loss'])
plt.plot(history_1.history['val_loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(['Training Loss','Validation Loss'],loc='upper right')

# Max acuuracy reached on Test or Validation Set : 
print(f"Max acuracy reached on Test or Validation Set : {max(history_1.history['val_accuracy'])}")
```
